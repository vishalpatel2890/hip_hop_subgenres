```python
import plotly as plt
import numpy as np
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot, plot
from plotly import tools
from plotly.tools import FigureFactory as ff
init_notebook_mode(connected = True)
```

```python
df_artists = pd.read_sql('artists', engine)
df_albums = pd.read_sql('albums', engine)
df_songs = pd.read_sql('songs', engine)
df_spotify = pd.read_sql('spotify_song_data', engine)
df_songs_merged = pd.merge(df_songs, df_albums, left_on = 'album_id', right_on = 'id')
len(df_spotify)
```

```python
df_songs_albums_merged = pd.merge(df_songs, df_albums, left_on = 'album_id', right_on = 'id')
df_songs_merged = pd.merge(df_songs_albums_merged, df_artists, left_on='artist_id', right_on = 'id')
df_songs_merged.columns = ['song_id', 'song_name', 'album_id', 'all_lyrics', 'artist_lyrics',
       'repetitiveness', 'topic', 'rhyme', 'album_id_x', 'album_name', 'artist_id', 'release_date',
        'artist_id_x', 'artist_name', 'decade']
df_songs_merged = df_songs_merged.drop(columns=['album_id_x', 'artist_id_x'])
df_songs_merged_spotify = pd.merge(df_songs_merged, df_spotify, left_on='song_id', right_on = 'song_id')

df_songs_merged['repetitiveness'] = df_songs['repetitiveness'].apply(lambda x: float(x))
df_songs_merged_spotify['repetitiveness'] = df_songs_merged_spotify['repetitiveness'].apply(lambda x: float(x))
df_songs_merged['count'] = 1
df_songs_merged_spotify['count'] = 1
```

```python
def create_topic_plot(df):
    data=[]
    for topic in [0,1,2,3,4]:
        years = [1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
        year_data = []
        years_appended =[]
        for x in years: 
            try: 
                n = (df_songs_merged.groupby(df.release_date.dt.year).topic.value_counts()[x].sort_index()[topic])/df.groupby(df_songs_merged.release_date.dt.year)['count'].sum()[x]
                year_data.append(n)
                years_appended.append(x)
            except:
                pass

        trace0 = go.Scatter(
                x = years_appended,
                y= year_data,
                line=dict(
                shape='spline',
            ),
            marker=dict(
            ),
            name=f'Topic {topic + 1}'
        )
        data.append(trace0)

    layout = dict(
        title = 'Topic Prevalance by Year',
        xaxis=dict(
            title='Year',
            titlefont=dict(
                family='Courier New, monospace',
                size=20,
                color='#000000'
            )
        ),
        yaxis=dict(
            title='Percentage of All Topics',
            titlefont=dict(
                family='Courier New, monospace',
                size=20
            )
        )
    )
#  
  
    fig = dict(data=data, layout=layout)
    plot(fig, filename='topic_prevalance.html')
    
    
```

```python
create_topic_plot(df_songs_merged)
```

```python
fig, ax = pyplot.subplots(figsize=(15,10))
ax.bar(['1980s', '1990s', '2000s', '2010s'],
           height=df_songs_merged.groupby('decade').mean()['repetitiveness'], 
           color=['#604860', '#ffdb58', '#F46D25', '#fb9fa4'])
ax.set_ylabel('Average Repetitiveness for Decade', fontsize=14)
ax.set_title('Repetitiveness By Decade', fontsize=20)
```

```python
def create_spotify_plot(df, spotify_key):
    data=[]
    metric = df_songs_merged_spotify.groupby(df_songs_merged_spotify.release_date.dt.year).mean()[spotify_key]

    trace0 = go.Scatter(
            x = list(metric.index),
            y= metric,
            line=dict(
            shape='spline',
        ),
        marker=dict(
        ),
        name=f'{spotify_key}'
    )
    data.append(trace0)

    layout = dict(
        title = f'{spotify_key} by Year',
        xaxis=dict(
            title='Year',
            titlefont=dict(
                family='Courier New, monospace',
                size=20,
                color='#000000'
            )
        ),
        yaxis=dict(
            title=f'{spotify_key}',
            titlefont=dict(
                family='Courier New, monospace',
                size=20
            )
        )
    )
#  
  
    fig = dict(data=data, layout=layout)
    plot(fig, filename=f'visuals/{spotify_key}.html')
```

```python
import tensorflow as tf
from tensorflow.contrib.tensorboard.plugins import projector

LOG_DIR = 'logs'
board_data_tf = tf.Variable(X, name='board_data_tf')

metadata = os.path.join(LOG_DIR, 'metadata.tsv')
with open(metadata, 'w') as metadata_file:
    for row in board_data_tf: # write to metadata file
        metadata_file.write('%s\n' % row)

with tf.Session() as sess:
    saver = tf.train.Saver([board_data_tf])
    sess.run(board_data_tf.initializer)
    saver.save(sess, os.path.join(LOG_DIR, 'board_data_tf.ckpt'))
    config = projector.ProjectorConfig()
    # One can add multiple embeddings.
    embedding = config.embeddings.add()
    embedding.tensor_name = board_data_tf.name
    # Link this tensor to its metadata file (e.g. labels).
    embedding.metadata_path = metadata
    # Saves a config file that TensorBoard will read during startup.
    projector.visualize_embeddings(tf.summary.FileWriter(LOG_DIR), config)
```
